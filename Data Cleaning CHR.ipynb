```python
import requests  # Import the requests library for making HTTP requests
import pandas as pd  # Import the pandas library for data manipulation
import re  # Import the re module for regular expressions
import numpy as np  # Import the numpy library for numerical operations
import json  # Import the json module for handling JSON data
import os

import configparser

config = configparser.ConfigParser()
config.read("my_config.ini")

pd.options.display.max_rows = None  # Display all rows of a DataFrame
pd.options.display.max_columns = None  # Display all columns of a DataFrame
pd.options.mode.chained_assignment = None
```

```python
year_list = [
    "2010",
    "2011",
    "2012",
    "2013",
    "2014",
    "2015",
    "2016",
    "2017",
    "2018",
    "2019",
    "2020",
    "2021",
    "2022",
    "2023",
    "2024",
]

df_list = []

for year in year_list:
    df = pd.read_csv(
        f"data/County Health Rankings/analytic_data{year}.csv",
        header=1,
        dtype={"fipscode": str, "statecode": str, "countycode": str},
    )

    headings_list = []

    for heading in df.columns:
        if re.match(r".*rawvalue$", heading):
            headings_list.append(heading)
        elif re.match(r".*other_data_[123]$", heading):
            headings_list.append(heading)

    headings_list.extend(["fipscode", "state", "year", "statecode", "countycode"])

    df = df[headings_list]

    df_list.append(df)
```

```python
chr_df = pd.concat(df_list, axis=0)
chr_df.head()
```
